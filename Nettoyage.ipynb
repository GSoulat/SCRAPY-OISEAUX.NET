```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np

df_scraping = pd.read_csv("WebCrawler\WebCrawler/oiseaux.csv")

df_scraping.shape
```

```python
df_scraping.head(5)
```

```python
df_oiseaux = pd.read_csv("WebCrawler\WebCrawler/oiseauxnet.csv")

df_oiseaux.shape
```

```python
df_oiseaux.head(5)
```

```python
df = df_scraping.merge(df_oiseaux, left_on='title', right_on='french_name',
          suffixes=('_left', '_right'))
```

```python
df = df.replace({np.nan: ''})
```

```python
df["envergure"] = df["envergure"].replace({":": ''}, regex=True)
df["poids"] = df["poids"].replace({":": ''}, regex=True)
df["taille"] = df["taille"].replace({":": ''}, regex=True)
```

```python
df.rename(columns={'id_bird': 'id'}, inplace=True)
```

```python
df.to_csv('OiseauxFini.csv', index = True)
df.head(5)
```
